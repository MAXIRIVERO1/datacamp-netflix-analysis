```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
```

```python
print(netflix_df)
```

```python
#What was the most frequent movie duration in the 90s?

#filtering movies between 1990 and 1999
movies = netflix_df[(netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] < 2000)]
print(movies)

#creating a histogram to visualize the most frequent movie duration
plt.hist(movies["duration"], 100)
plt.show()

#mode returns the most common value to verify the result
duration = movies["duration"].mode()[0]
print(f"The most frequent movie duration in the 90s is {duration} minutes")

#anser: The most frequent movie duration in the 90s is 94
```

```python
#Quantity of short action movies from the 90s

#Subsetting the action movies from the dataframe
action = movies[movies["genre"] == "Action"]
print(action)

#Iterating action movies to find short durations
short_movie_count = 0
for _, row in action.iterrows() :
    if row["duration"] < 90 :
        short_movie_count = short_movie_count + 1

print(f"There are {short_movie_count} short movies from the 90s")

#There are 7 short movies from the 90s

    
```
